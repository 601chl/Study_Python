#### 판다스
- 판다스는 구조화된 데이터 형식을 제공. 시리즈는 1차원 배열, 데이터프레임은 2채원 배열
- 라이브러리는 여러 종류의 class와 다양한 내장 함수로 구성.시리즈(위에서 아래로, 하나의 열, 컬럼이 하나.)와 데이터 프레임(행렬,/여러개의 컬럼;시리즈가 모여서 프레임이 됨.배열이 여러개 모여있음,2차원배열)은 대표적인 클래스 객체임.
- 시리즈 인덱스는 데이터 값과 일대일 대응. 파이썬 딕셔너리와 비슷한 구조. pandas.Series(딕셔너리)
- Contnets, 판다스에 들어있는 클래스
 + Series(변환, 인덱스 구조, 원소 선택),컬럼이름없음 [키 - 인덱스, 값 - 내용]
 + DataFrame(변환, 행인덱스/열이름 지정, 삭제, 선택, 추가, 변경, 전치(행,열변환), 인덱스 활용), 컬럼이름있음      
      *행 - row, 레코드(record), 관측값(observation)      
      *열 - column, 공통의 속성을 갖는 일련의 데이터, 속성, 범주, 변수(variable)로 활용

#### 데이터프레임
- 데이터프레임은 2차원 배열, R의 테이터 프레임에서 유래.(2차원은 직선으로 나눔, 3차원은 평면으로 나눔, 4차원이상은 수학적으로 나눔)
- 데이터프레임의 열은 각각 시리즈 개체.
- 시리즈를 열벡터(열로되어있는 스칼라(숫자, 하나의 원소)가 모여진 벡터)라고 하면 데이터프레임은 여러개의 열벡터들이 같은 행 인덱스를 기준으로 줄지어 결합된 2차원 벡터 또는 행렬.(한 행에는 열백터에 배치되는 원소들)

>
출처 : https://rfriend.tistory.com/144

- 선대형수학에서 열 벡터(m*1 행렬)는 m 원소들의 단일 열 행렬
- 행 벡터(1*m 행렬)은 m원소들의 단일 행 행렬.
- 리스트, 딕셔너리, ndarry 등 다양한 데이터로부터 생성
- 반대로 리스트, 딕셔너리, ndarray 등으로 변환될 수 있음.

```python
#시리즈 만들기
import pandas as pd
st1= pd.Series({'국어':100,'영어':80,'수학':90})
st2= pd.Series({'수학':80,'국어':90,'영어':80})
print(st1)
print(st2)
```

```python
#시리즈끼리 연선
add = st1 + st2   #속성이 같은것끼리 더해짐.
sub = st1 - st2
mul = st1 * st2
div = round((st1 / st2),2) #반올림처리 소숫점이하자리 설정 2
print(add)
```

```python
#위의 연산된 시리즈로 데이터프레임 만들 수 있음.
#데이터프레임생성
df = pd.concat([add,sub,mul,div],axis=1)  #열방향axis=1
```
